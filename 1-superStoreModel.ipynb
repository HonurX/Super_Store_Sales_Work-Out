ADDING SEASON COLUMN FOR SALE INDEX


```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")
```

```python
def get_seaoson(date):
    month =date.month
    if month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    elif month in [9,10,11]:
        return "Autumn"
    else:
        return "Winter"
        
df["Order_Season"] = df["Order Date"].apply(get_seaoson)       
```

```python
column_order= ["Order Date" , "Order_Season" ]+ [ col for col in df.columns if col not in ["Order Date" , "Order Season"]]
df= df[column_order]
```
